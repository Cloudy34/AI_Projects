```python
print(df.shape)

df.dropna(inplace =True)
print(df.shape)
```

```python
print(df.dtypes)
```

```python
df
```

```python
del df['link']
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df['Neighborhood'] = le.fit_transform(df['Neighborhood'])
df['Agency'] = le.fit_transform(df['Agency'])
df.head(3)

```

```python
df['Price'] = df['Price'].str.replace('KSh','')
df['Price'] = df['Price'].str.replace(',','')

df['Price'] = df['Price'].astype(int)

df.head(5)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

df[['Agency','Neighborhood','sq_mtrs']] = sc.fit_transform(df[['Agency','Neighborhood','sq_mtrs']])
df.head(5)
```

```python
df.describe()
```

```python
df = df.sample(frac = 1).reset_index(drop = True)
```

```python
import random

ts = (random.randint(5,50)) /100
print(ts)

rs = random.randint(1,50)
print(rs)
```

```python
from sklearn.model_selection import train_test_split as tts

x = df.drop(['Price'],axis = 1)
y = df['Price']
```

```python
x_train,x_test,y_train,y_test = tts(x,y, test_size = ts, random_state = rs)
print(x_train.shape, x_test.shape)
print(y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
pred_lr = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error as mse

print(mse(y_test,pred_lr))
```
